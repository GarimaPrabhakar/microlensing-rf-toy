```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
t = np.linspace(-50, 50)

for i in range(10):
    u0 = np.random.uniform(0.1, 1)
    tE = np.random.uniform(5, 100)
    # u0 = 0.5
# u0 = 1
    tE = 25
    u0 = 0.5
    t0 = np.random.uniform(2*min(t), 2*max(t))

    u = (u0**2 + ((t - t0)/tE)**2)**0.5
    A = (u + 2)/(u*(u + 4)**0.5)
    A = A/min(A)
    # label=f"u0={round(u0, 2)}, tE={round(tE)}
    plt.plot(t, A, label=f"t0={round(t0)}")
plt.legend()
```
