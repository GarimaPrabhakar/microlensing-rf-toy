```python
def pspl_model(mjd, f, u0, tE, t0):
    x = mjd
    t = x - np.min(x) - (np.max(x) - np.min(x))/2
    u = (u0**2 + ((t - t0)/tE)**2)**0.5
    A = (u + 2)/(u*(u + 4)**0.5)
    A = A/min(A)
    
    return mjd, A*f
```
